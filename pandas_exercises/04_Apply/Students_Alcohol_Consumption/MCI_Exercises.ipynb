### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
legal = lambda x : True if x > 17 else False
```

```python
legal(17)
```

```python
df['legal_drinker'] = df.age.apply(legal)
df.head()
```

```python
# Other way

df['legal_drinker_2'] = np.where(df.age > 17, True, False)
df.head(10)
```

```python
(df.legal_drinker != df.legal_drinker_2).sum()
```
